# Hypothesis Testing

# Titanic Dataset

```python
from scipy.stats import ttest_ind

male = Titanic_df[Titanic_df['Sex'] =='male']['Survived']
female = Titanic_df[Titanic_df['Sex'] == 'female']['Survived']

t_stat, p_value = ttest_ind(female, male)

print("T-stat:", t_stat)
print("P-value:", p_value)
```

```python
from scipy.stats import f_oneway

group1 = Titanic_df[Titanic_df['Pclass'] == 1]['Survived']
group2 = Titanic_df[Titanic_df['Pclass'] == 2]['Survived']
group3 = Titanic_df[Titanic_df['Pclass'] == 3]['Survived']

f_stat, p_value = f_oneway(group1, group2, group3)

print(p_value)
```

# Housing Dataset

Hypothesis Test 1:

Questions

Does house area affect price?

Hypotheses:

H₀: No relationship between area and price
H₁: Area affects price

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(Housing_df['area'],Housing_df['price'])

print("Correlation:", corr)
print("P-value:", p_value)
```

```python
import matplotlib.pyplot as plt

plt.scatter(Housing_df['area'], Housing_df['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

Hypothesis Test 2: Preferred Area vs Price

Question:

Do houses in preferred areas cost more?

```python
from scipy.stats import ttest_ind

pref = Housing_df[Housing_df['prefarea'] == 'yes']['price']
not_pref = Housing_df[Housing_df['prefarea'] == 'no']['price']

t_stat, p_value = ttest_ind(pref, not_pref)

print(p_value)
```

There is a statistically significant difference in house prices between preferred and non-preferred areas.

Hypothesis Test 3: Furnishing Status (ANOVA)
Question:

Does furnishing status affect price?


```python
from scipy.stats import f_oneway

furnished = Housing_df[Housing_df['furnishingstatus'] == 'furnished']['price']
semi = Housing_df[Housing_df['furnishingstatus'] == 'semi-furnished']['price']
unfurnished = Housing_df[Housing_df['furnishingstatus'] == 'unfurnished']['price']

f_stat, p_value = f_oneway(furnished, semi, unfurnished)

print("P-value:", p_value)
```

Hypothesis Test 5: Bedrooms vs Price

Question:

Do more bedrooms increase price?

```python
corr, p_value = pearsonr(Housing_df['bedrooms'], Housing_df['price'])

print(corr, p_value)
```
